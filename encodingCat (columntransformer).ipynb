```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,RobustScaler
```

```python
df = pd.read_csv("C:/Users/user/Desktop/kaggle competetion/Churn_Modelling.csv")
df.shape
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop("Exited",axis =1),
                                                 df.Exited,test_size=0.2,random_state=0)

```

```python
num_cols = [col for col in x_train.columns if x_train[col].dtypes!='O']
num_cols
```

```python
cat_cols = [col for col in x_train.columns if x_train[col].dtypes == "O"]
cat_cols
```

## scaling numerical

```python
r = RobustScaler()
r.fit(x_train[num_cols])
```

```python
x_train_num_scaled = r.transform(x_train[num_cols])
```

```python
x_test_num_scaled = r.transform(x_test[num_cols])
```

## onehotencoding

```python
ohe = OneHotEncoder(handle_unknown="ignore",sparse = False)
ohe.fit(x_train[cat_cols])
```

```python
x_train_cat_encoded = ohe.transform(x_train[cat_cols])
pd.DataFrame(x_train_cat_encoded)
```

```python
x_test_cat_encoded = ohe.transform(x_test[cat_cols])
```

```python
pd.DataFrame(np.concatenate((x_train_num_scaled,x_train_cat_encoded),axis =1))
```

## column transformer

```python
# ColumnTransformer epects a list of tuples
```

```python
ct = ColumnTransformer([
    ("step1",RobustScaler(),num_cols),
    ("step2",OneHotEncoder(sparse=False,handle_unknown="ignore"),cat_cols),
],remainder = "drop")
```

```python
ct.fit(x_train)
```

```python
ct.transform(x_train)
```

```python
ct.transform(x_test)
```

```python
pd.DataFrame(ct.transform(x_train))
#the column names are dropped now lets get the column names
```

```python
ct.transformers_
```

```python
ct.transformers_[1][1].get_feature_names()
#these are the column names given by sklearn for categorigal columns(step 2)
```
